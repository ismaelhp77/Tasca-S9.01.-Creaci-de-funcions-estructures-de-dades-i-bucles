Existeixen diverses unitats de temperatura utilitzades en diferents contextos i regions. Les m√©s comunes s√≥n Celsius (¬∞C), Fahrenheit (¬∞F) i Kelvin (K). Tamb√© existeixen altres unitats com Rankine (¬∞Ra) i R√©aumur (¬∞Re).

Selecciona almenys 2 conversors, de tal manera que en introduir una temperatura retorni, com a m√≠nim, dues conversions, de manera que es puguin guardar (recorda que un print() no es pot guardar mai).

Consell: Intenta validar les dades pr√®viament, perqu√® envi√Ø un missatge d‚Äôadvert√®ncia si les dades introduides per l‚Äôusuari/a no estan en el format adequat.

```python
def get_temp():
    temp= float(input("Introduzca  una temperatura "))
    return temp
```

```python
def get_unidad():
    listavalidas =["c","k","f","celsius","kelvin","fahrenheit"]
    unidad= str(input("Introduzca la unidad de medida base de su temperatura"))
    unidad = unidad.lower()
    if unidad in listavalidas :
     return unidad
    else:
       print(f"{unidad} no es una unidad de medida valida")
```

```python
def Conversor_Celsius(temp):
    
    tempk = round((temp + 273.15),2)
    tempf =  round(((temp * 9/5) + 32),2)
    print(f"Tu temperatura de {temp}¬∞C es {tempk}¬∞K y  {tempf}¬∞F")
    return temp,tempk,tempf
```

```python
def Conversor_Kelvin(temp):
    
    tempc =  round((temp - 273.15),2)
    tempf =  round((((temp - 273.15)*9/5) + 32),2)
    print(f"Tu temperatura de {temp}¬∞K es {tempc}¬∞C y  {tempf}¬∞F")
    return tempc,temp,tempf
```

```python
def Conversor_Fahrenheit(temp):
    
    tempk =  round((((temp - 32)*9/5) + 273.15),2)
    tempc =  round(((temp - 32)*9/5),2)
    print(f"Tu temperatura de {temp}¬∞F es {tempk}¬∞K y  {tempc}¬∞C")
    return tempc,tempk,temp
```

```python
def conversor_temp():
   temp=get_temp()
   unidad=get_unidad()
   if unidad == "c" or unidad == "celsius":
     if temp  < -273.15:
       print(f"{temp}¬∞C no es posible fisicamente")
     else:
      Conversor_Celsius(temp)
   elif unidad == "k" or unidad == "kelvin":
           if temp  < 0:
              print(f"{temp}¬∞K no es posible fisicamente")
           else:
             Conversor_Kelvin(temp)
   elif unidad == "f" or unidad == "fahrenheit" :
        if temp  < -459.67:
         print(f"{temp}¬∞F no es posible fisicamente")
        else:
           Conversor_Fahrenheit(temp)
   
```

```python
def invertir_dicc(diccionario):
 diccionario_invertido = dict()

 for clave, valor in diccionario.items():
    if valor in diccionario_invertido:
        diccionario_invertido[valor].append(clave)
    else:
        diccionario_invertido[valor] = [clave]

 print(diccionario_invertido)        
```

Per a solucionar-ho ens ha encarregat una funci√≥ de Python que generi contrasenyes m√©s segures. La funci√≥ ha de dependre dels seg√ºents par√†metres:

longitud (int): Longitud de la contrasenya
majuscules (bool = True): Si hi ha d‚Äôapar√®ixer maj√∫scules
minuscules (bool = True): Si hi ha d‚Äôapar√®ixer min√∫scules
numeros (bool = True): Si hi ha d‚Äôapar√®ixer n√∫meros
signes (bool = False): Si hi ha d‚Äôapar√®ixer car√†cters especials (,-$? o similars)
Aix√≠ doncs, si executem la funci√≥ de la seg√ºent manera:

Haur√≠em d‚Äôobtenir un output (que hem de poder guardar) de l‚Äôestil:

9Er,5Vn8P$

Assegura‚Äôt que es compleixin tots els criteris, i que aquestes contrasenyes siguin realment aleat√≤ries.

(EXTRA) Explora com podr√≠em fer que la funci√≥ copi√©s la contrasenya autom√†ticament al porta papers de l‚Äôordinador (com si l‚Äôhagu√©ssim seleccionat i fet ctrl+copy).

```python

import string

import secrets

import pyperclip



# Utilizo la libreria "sectrets" en vez de "random ya que se supone que es mas segura."
```

```python


def validar_contrase√±a(contrase√±a, validacion):
    checks = {
        "long": len(contrase√±a) == validacion.get("long", len(contrase√±a)),
        
        "mayus": any(c.isupper() for c in contrase√±a) 
                 if validacion.get("requerir_mayus", False) else True,
        
        "minus": any(c.islower() for c in contrase√±a) 
                 if validacion.get("requerir_minus", False) else True,
        
        "num": any(c.isdigit() for c in contrase√±a) 
               if validacion.get("requerir_num", False) else True,
        
        "signos": any(c in string.punctuation for c in contrase√±a) 
                  if validacion.get("requerir_signo", False) else True
    }

    return all(checks.values())
```

version sin extra

```python
def crear_contrase√±a(long=10,mayus=True,minus=True,num=True,signos=False):
 validacion={
 "longitud": long,
 "requerir_mayus":mayus,
 "requerir_minus":minus,
 "requerir_num":num,
 "requerir_signo":signos
 }
 
 characters=string.ascii_letters + string.digits + string.punctuation 
 contrase√±a= ''.join(secrets.choice(characters) for _ in range(long))
 
 valida=validar_contrase√±a(contrase√±a,validacion)

 if valida == True :
  return contrase√±a
 else:
      
  return crear_contrase√±a(long, mayus, minus, num, signos)

 
 

 




```
